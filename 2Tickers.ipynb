```python
import matplotlib.pyplot as plt
import yfinance as yf

# Download data
spy_data = yf.download('SPY', start='2022-01-01', end='2022-09-21')
vix_data = yf.download('^VIX', start='2022-01-01', end='2022-09-21')

# Create subplot
fig, ax1 = plt.subplots()

# Plot SPY on the first y-axis
ax1.plot(spy_data.index, spy_data['Close'], label='SPY', color='b')
ax1.set_xlabel('Date')
ax1.set_ylabel('SPY Close Price', color='b')

# Create second y-axis
ax2 = ax1.twinx()
ax2.plot(vix_data.index, vix_data['Close'], label='VIX', color='r')
ax2.set_ylabel('VIX Close Price', color='r')

# Show plot
plt.title('SPY and VIX on the Same Chart')
fig.tight_layout()
plt.show()
```
